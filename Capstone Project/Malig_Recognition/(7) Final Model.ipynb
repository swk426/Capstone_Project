# Final Model Tuning

The object of this module is to further tuning of the final model to check which model can bring highest possible accuracy.

## Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
from datetime import datetime
import random

from tensorflow.keras import models
from tensorflow.keras import layers
from tensorflow.keras.models import load_model
from tensorflow.keras import regularizers
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

from sklearn.metrics import mean_squared_error
from sklearn.metrics import confusion_matrix

from malig_data import *
import util as ut
```

In previous model, it seemed like it was undertrained with higher validation accuracy. For final tuning, regularizers parameter is changed from 0.005 to 0.01. Patient for early stop is changed to 150 from 50. Drop out layer rate has changed from 0.3 to 0.2. Ephochs is increased from 200 to 250. In order to mitigate the time consumption, batch is changed to 64 from 32.

### First Model Tuning

This model will have one additional hidden layer.

```python
def CNN_model(x, y, val_x, val_y, filepath, epochs=150, batch_size=32):
        """ CNN model function that returns the model and the history.
        x: train image
        y: train target
        val_x: test image
        val_y: test target
        filepath: where the model to be saved
        epochs: default at 150
        batch_size: default at 32"""
    start_time = datetime.now()
    random.seed(123)
    cnn_model = models.Sequential()
    cnn_model.add(layers.Conv2D(32, (3, 3), activation='relu',
                            input_shape=(64,64,3))) #1st Hidden Layer
    cnn_model.add(layers.MaxPooling2D((2, 2))) #Maxpooling to 1st Layer

    cnn_model.add(layers.Conv2D(32, (3, 3), activation='relu')) #2nd Hidden Layer
    cnn_model.add(layers.MaxPooling2D((2,2))) #Maxpooling to 2nd Layer
    cnn_model.add(layers.Dropout(.2)) #Dropout layer to 2nd layer
    
    cnn_model.add(layers.Conv2D(32, (3, 3), activation='relu', kernel_regularizer = regularizers.l2(0.01))) #3rd Hidden Layer
    cnn_model.add(layers.MaxPooling2D((2, 2))) #Maxpooling to 3rd Layer
    cnn_model.add(layers.Dropout(.2)) #Dropout layer to 3rd layer

    cnn_model.add(layers.Conv2D(64, (3, 3), activation='relu', kernel_regularizer = regularizers.l2(0.01)))#4th Hidden Layer
    cnn_model.add(layers.MaxPooling2D((2, 2))) #Maxpooling to 4th Layer
    cnn_model.add(layers.Dropout(.2)) #Dropout layer to 4thlayer

    cnn_model.add(layers.Flatten()) #Flattening Layer
    cnn_model.add(layers.Dense(64, activation='relu', kernel_regularizer = regularizers.l2(0.01)))#5th Hidden Layer
    cnn_model.add(layers.Dense(1, activation='sigmoid')) #Output Layer

    cnn_model.compile(loss='binary_crossentropy',
                  optimizer="adam", # For CNN model we will use 'adam' as an optimizer which is most commonly used optimizer
                  metrics=['accuracy'])

    cnn_early_stopping = [EarlyStopping(monitor='val_loss', patience=150),
                      ModelCheckpoint(filepath=filepath,
                                     monitor='val_loss',
                                     save_best_only=True)]

    cnn_history = cnn_model.fit(x, y,
                        epochs=epochs,
                        batch_size=batch_size,
                        validation_data=(val_x, val_y),
                        callbacks=cnn_early_stopping)

    end_time = datetime.now()

    print('Time elapsed', end_time - start_time)
    return cnn_model, cnn_history
```

```python
filepath = '../models/final_model.h5'
final_cnn_model, final_cnn_history = CNN_model(adasyn_images, adasyn_labels, 
                                               test_images, test_y,
                                               filepath=filepath,
                                               epochs=250,    #epochs number is increased to 250 to avoid 
                                               batch_size=64) #Changed the batch size to 64 from 32
```

```python
final_cnn_saved_model = load_model(filepath)

final_cnn_results_train = final_cnn_saved_model.evaluate(adasyn_images, adasyn_labels)
print(f'Training Loss: {final_cnn_results_train[0]:.3} \nTraining Accuracy: {final_cnn_results_train[1]:.3}')

print('----------')

final_cnn_results_test = final_cnn_saved_model.evaluate(val_images, val_y)
print(f'Test Loss: {final_cnn_results_test[0]:.3} \nTest Accuracy: {final_cnn_results_test[1]:.3}')
```

```python
title = "../reports/Final CNN Model: Iteration of Loss Graph"
ut.training_results_Loss(final_cnn_history, title=title)
title = '../reports/Final CNN Model: Iteration of Accuracy Graph'
ut.training_results_Accuracy(final_cnn_history, title=title)
```

```python
index=["Actual Malig", "Actual Benign"]
columns=["Predicted Malig", "Predicted Benign"]
final_cnn_predictions = final_cnn_saved_model.predict_classes(val_images)
final_cnn_cm = confusion_matrix(val_y, final_cnn_predictions, labels=[0,1])
ut.cm_df(final_cnn_cm, index, columns)
```

It seems like underfitting issue has been resolved and accuracy is improved a little.
